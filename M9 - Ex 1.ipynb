### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_raw['munResNome'].value_counts()
```

Município de Castanheiras teve menos ocorrências.


```python
#idade media das mães no município de Castanheiras.
media = sinasc_raw[sinasc_raw['munResNome'] == 'Castanheiras']['IDADEMAE'].mean()
print("Idade média em Castanheira:",media)
#idade mínima das mães no município de Castanheiras.
minima = sinasc_raw[sinasc_raw['munResNome'] == 'Castanheiras']['IDADEMAE'].min()
print("Idade minima em Castanheira:",minima)
maxima = sinasc_raw[sinasc_raw['munResNome'] == 'Castanheiras']['IDADEMAE'].max()
print("Idade máxima em Castanheira:",maxima)
```

No Munícipio de Castanheiras as idades das mães possuiem uma grande amplitude.

```python
#idade media das mães no município de Castanheiras.
media = sinasc_raw[sinasc_raw['munResNome'] == 'Castanheiras']['IDADEPAI'].mean()
print("Idade média em Castanheira:",media)
#idade mínima das mães no município de Castanheiras.
minima = sinasc_raw[sinasc_raw['munResNome'] == 'Castanheiras']['IDADEPAI'].min()
print("Idade minima em Castanheira:",minima)
maxima = sinasc_raw[sinasc_raw['munResNome'] == 'Castanheiras']['IDADEPAI'].max()
print("Idade máxima em Castanheira:",maxima)
```

A amplitude das idades dos pais é maior que amplitude na idade das mães. Percebe-se também que existem muitos pais com idade maior que a das mães de acordo com a média das idades.

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
mun = sinasc_raw[['munResNome','DTNASC','QTDFILVIVO','IDADEMAE','IDADEPAI']]
mun
```

```python
data1 = pd.to_datetime(mun['DTNASC'])
mun['mês'] = pd.DataFrame(data1.dt.month)
mun
```

```python
mun[mun['mês'] == 3].groupby('munResNome')['DTNASC'].count().sort_values(ascending=False)
```

O município de Porto Velho possui o maior números de nascimento no mês de março.

```python
media_filho_vivo = mun[mun['munResNome'] == 'Porto Velho']['QTDFILVIVO'].mean()
print(media_filho_vivo)
minimo_filho_vivo = mun[mun['munResNome'] == 'Porto Velho']['QTDFILVIVO'].min()
print(minimo_filho_vivo)
maximo_filho_vivo = mun[mun['munResNome'] == 'Porto Velho']['QTDFILVIVO'].max()
print(maximo_filho_vivo)


```

```python
media_idade_pai = mun[mun['munResNome'] == 'Porto Velho']['IDADEPAI'].mean()
print(media_idade_pai)
minimo_idade_pai = mun[mun['munResNome'] == 'Porto Velho']['IDADEPAI'].min()
print(minimo_idade_pai)
maximo_idade_pai = mun[mun['munResNome'] == 'Porto Velho']['IDADEPAI'].max()
print(maximo_idade_pai)
```

A amplitude da idade dos pais nesse municipio é maior que no município de Castanheira. A media das idades também é maior e isso 
se deve ao fato de que nesse lugar há homens mais velhos registrados como pais.
