#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
#importando pacotes
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#carregando a base
base = pd.read_csv('b1abb33c4f.csv')
base
```

```python
#selecionando colunas
base_select = base[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE','CONSULTAS']]
base_select
```

```python
#Tratando dados ausentes
base_select['APGAR1'].fillna(9,inplace = True )
base_select['APGAR5'].fillna(9,inplace = True )
base_select['SEXO'].fillna('Masculino',inplace = True )
base_select['GESTACAO'].fillna('37 a 41 semanas',inplace = True )
base_select['ESTCIVMAE'].fillna(9,inplace = True )
```

```python
base_select.isnull().sum()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(15, 8)) 
fig, axes = plt.subplots(2, 3) 

sns.countplot(ax = axes[0, 0], x='APGAR1', data=base_select) 
sns.countplot(ax = axes[0, 1], x='APGAR5', data=base_select)
sns.countplot(ax = axes[0, 2], x='SEXO', data=base_select)
sns.countplot(ax = axes[1, 0], x='GESTACAO', data=base_select)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=base_select)
axes[1,0].tick_params(axis="x", rotation = 90)


#ajustar o espaçamento
plt.subplots_adjust(wspace=0.25, hspace=0.25)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.rc('figure', figsize=(4, 5))
sns.barplot(x='CONSULTAS', y='APGAR5', data=base_select)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 10))
fig, axes = plt.subplots(2,2)
sns.countplot(ax = axes[0,0], x = 'GESTACAO', data = base_select)
sns.countplot(ax = axes[0,1], x = 'ESTCIVMAE', data = base_select)
sns.countplot(ax = axes[1,0], x = 'SEXO', data = base_select)
sns.countplot(ax = axes[1,1], x = 'CONSULTAS', data = base_select)
axes[0,0].tick_params(axis="x", rotation = 90)
plt.subplots_adjust(wspace=0.7, hspace=0.7)
```
